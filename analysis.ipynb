# TSLA vs BYD Stock Analysis
Target: Compare 1-year risk for retail investors
Data source: Yahoo Finance

```python
import yfinance as yf
import pandas as pd
import plotly.express as px
from datetime import datetime, timedelta

end = datetime.today()
start = end - timedelta(days=365)

tsla = yf.download("TSLA", start=start, end=end)["Adj Close"]
byd = yf.download("1211.HK", start=start, end=end)["Adj Close"]
df = pd.DataFrame({"Tesla": tsla, "BYD": byd}).dropna()

returns = df.pct_change().dropna()

fig = px.line(df, title="Stock Price Trend")
fig.show()

print(f"Tesla Annual Volatility: {returns['Tesla'].std() * (252**0.5) * 100:.2f}%")
print(f"BYD Annual Volatility: {returns['BYD'].std() * (252**0.5) * 100:.2f}%")
```
